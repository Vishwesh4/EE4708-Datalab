# Time Series :
    A sequential set of data points measured over time.  
    x(t), t- variable that repersents the elapsed time and t=0,1,2,3,....
### How Time Series is special? 
    It is time dependent. So basic assumption of regression/classification model that the observations are independent doesn't hold in this. 

### Examples:
    Daily temperature
    monthly rain fall
    daily closing price of company stock
    continuous monitoring of a person's heart rate. 

# Auto Regression
    A linear model that uses the dependent relationship between an observation and some number of lagged observations.
    
\begin{equation*}
X_t=\phi_1 X_{t-1}+\phi_2 X_{t-1}....\phi_p X_{t-p}+ w_t \\
X_t= \sum_{i=1}^p \phi_i X_{t-i} + w_t \\
w_t : white noise \sim wn(0,\sigma _w ^2)
\end{equation*}

p : hyper parameter that repersents the length of the direct look back in the series

## Q1) Genarate a time series of 1000 data points with following AR(p) models.
1. AR(0) with $\sigma_w^2=0.3$
2. AR(1) with $\sigma_w^2=0.3 , \phi_1=0.3$
3. AR(1) with $\sigma_w^2=0.3 , \phi_1=0.9$
4. AR(2) with $\sigma_w^2=0.3 , \phi_1=0.4, \phi_2=0.4$
5. AR(2) with $\sigma_w^2=0.3 , \phi_1=0.4, \phi_2=-0.4$


Visualize all 5 time series using subplots and give proper legends and titles. 

```python
import numpy as np
from matplotlib import pyplot as plt
#1
Ar_0 = np.zeros(1000)
Ar_1 = np.zeros(1000)
Ar_2 = np.zeros(1000)
Ar_3 = np.zeros(1000)
Ar_4 = np.zeros(1000)


#initialization
#AR_0
for i in range(1000):
    error_generate = np.random.randn()*np.sqrt(0.3)
    Ar_0[i] = error_generate
#AR_1
Ar_1[0] = np.random.randn()*np.sqrt(0.3)
Ar_2[0] = np.random.randn()*np.sqrt(0.3)
for i in range(1,1000):
    error_generate = np.random.randn(2)*np.sqrt(0.3)
    Ar_1[i] = 0.3*Ar_1[i-1]+error_generate[0]    
    Ar_2[i] = 0.9*Ar_2[i-1]+error_generate[1]    
#AR_2
Ar_3[0] = np.random.randn()*np.sqrt(0.3)
Ar_4[0] = np.random.randn()*np.sqrt(0.3)
Ar_3[1] = 0.4*Ar_3[0]+np.random.randn()*np.sqrt(0.3)
Ar_4[1] = 0.4*Ar_4[0]+np.random.randn()*np.sqrt(0.3)
for i in range(2,1000):
    error_generate = np.random.randn(2)*np.sqrt(0.3)
    Ar_3[i] = 0.4*Ar_3[i-1]+0.4*Ar_3[i-2]+error_generate[0]    
    Ar_4[i] = 0.4*Ar_4[i-1]-0.4*Ar_4[i-1]+error_generate[1]    
```

```python
fig = plt.figure(figsize = [10,20])
plt.subplot(5,1,1)
plt.plot(np.arange(1000),Ar_0)
plt.title("AR(0) with  𝜎2𝑤=0.3")

plt.subplot(5,1,2)
plt.plot(np.arange(1000),Ar_1)
plt.title("AR(1) with  𝜎2𝑤=0.3,𝜙1=0.3")

plt.subplot(5,1,3)
plt.plot(np.arange(1000),Ar_2)
plt.title("AR(1) with  𝜎2𝑤=0.3,𝜙1=0.9")

plt.subplot(5,1,4)
plt.plot(np.arange(1000),Ar_3)
plt.title("AR(2) with  𝜎2𝑤=0.3,𝜙1=0.4,𝜙2=0.4")

plt.subplot(5,1,5)
plt.plot(np.arange(1000),Ar_4)
plt.title("AR(2) with  𝜎2𝑤=0.3,𝜙1=0.4,𝜙2=−0.4")

plt.show()
```

## Q3) Generate AR(1)  time series of 1000 data points with following parameters and visualize their autocovariance. 
1. AR(1) with $\sigma_w^2=0.3 , \phi_1=0.1$
2. AR(1) with $\sigma_w^2=0.3 , \phi_1=0.8$
3. AR(1) with $\sigma_w^2=0.3 , \phi_1=-0.4$

## Auto covariance: 
 
$Cov(X_t, X_{t-k})=E[(X_t - \mu)(X_{t-k} -\mu)]$ <br>

$ Cov(X_t, X_{t-k})=\frac {1}{(N-1)} \sum_{i=0}^{N-k} ( x_i - \mu )( x_{i+k} - \mu )$ <br>

To identify the dominant periods in the time series.


```python
#Generate AR(1) time series
#AR_1
Ar_1 = np.zeros(1000)
Ar_2 = np.zeros(1000)
Ar_3 = np.zeros(1000)


Ar_1[0] = np.random.randn()*np.sqrt(0.3)
Ar_2[0] = np.random.randn()*np.sqrt(0.3)
Ar_3[0] = np.random.randn()*np.sqrt(0.3)


for i in range(1,1000):
    error_generate = np.random.randn(3)*np.sqrt(0.3)
    Ar_1[i] = 0.1*Ar_1[i-1]+error_generate[0]    
    Ar_2[i] = 0.8*Ar_2[i-1]+error_generate[1]  
    Ar_3[i] = -0.4*Ar_2[i-1]+error_generate[2]  

#Auto covariabce
def autocovariance(X,N):
    mean = np.mean(X)
    autocovariance = []
    for lag in range(N):
        summ = 0
        for i in range(len(X)-lag):
            summ+= (X[i]-mean)*(X[i+lag]-mean)
        autocovariance.append(summ/(len(X)-1))
    return autocovariance 
N = 350
autovariance_1 = autocovariance(Ar_1,N)
autovariance_2 = autocovariance(Ar_2,N)
autovariance_3 = autocovariance(Ar_3,N)
```

```python
fig = plt.figure(figsize = [10,15])
plt.subplot(3,1,1)
plt.plot(np.arange(N),autovariance_1)
plt.title("Autocovariance AR(1) p=0.1")
plt.xlabel("lag")
plt.ylabel("autocovariance")

plt.subplot(3,1,2)
plt.plot(np.arange(N),autovariance_2)
plt.title("Autocovariance AR(1) p=0.8")
plt.xlabel("lag")
plt.ylabel("autocovariance")


plt.subplot(3,1,3)
plt.plot(np.arange(N),autovariance_3)
plt.title("Autocovariance AR(1) p=-0.4")
plt.xlabel("lag")
plt.ylabel("autocovariance")
```

## Q4) Given an AR(1) model try to find its $\phi$. 
Dataset : ar_1.csv <br>
Noise  : $W_t \sim N(0,0.3)$ <br>
$X_t=\phi_1 X_{t-1}+ w_t$ <br>
Compute the $\phi$


```python
import pandas as pd
df = pd.read_csv("ar_1.csv")
df['X_t-1'] = df['Data'].shift(1)
df.fillna(0,inplace=True)
#Fitting linear model for Y as X_t and X as X_t-1
Y = df['Data'].values
X = df['X_t-1'].values
#(X.T*X)^-1*(X.T*Y)
w = np.dot(X.T,Y)/(np.dot(X.T,X))
print("Phi:{}".format(w))
```

## Q5) Try to fit and forcast time series for a given real time data set using inbuilt ar library. 
Dataset : Electricity_comsumption_hourly.csv
Predict last 12 data points and compare with the actual data point. 

Train anf test data :

X=data['DEOK_MW'].values <br>
train, test = X[1:len(X)-12], X[len(X)-12:]

Inbuilt AR function : from statsmodels.tsa.ar_model import AR

```python
from statsmodels.tsa.ar_model import AR
df = pd.read_csv("Electricity_comsumption_hourly.csv")
X=df['DEOK_MW'].values 
train, test = X[1:len(X)-12], X[len(X)-12:]
```

```python
pd.plotting.autocorrelation_plot(df['DEOK_MW'])
plt.show()
```

```python
plt.plot(train)
plt.show()
```

```python
model = AR(train)
model_fitted = model.fit()
```

```python
print('lag value: {}'.format(model_fitted.k_ar))
print('coefficients are:\n {}'.format(model_fitted.params))
```

```python
# make predictions 
predictions = model_fitted.predict(start=len(train), end=len(train)+ 11)
print("Predictions:{}".format(predictions))
print("True values:{}".format(test))

plt.plot(np.arange(12),predictions)
plt.plot(np.arange(12),test)
plt.legend(["Predictions","True values"])
plt.show()
```
