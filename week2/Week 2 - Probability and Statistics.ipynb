### Basic moments of one-dimensional data

```python
import numpy as np
data1 = np.array([1,2,3,4,5])
def mean(data1):
    return(float(sum(data1))/len(data1))
def variance(data1):
    summed = 0
    for i in range(len(data1)):
        summed+=(data1[i]-mean(data1))**2
    return summed/len(data1)
def stddev(data1):
    return variance(data1)**(0.5)
def median(data1):
    data1.sort()
    n = len(data1)
    if (n+1)%2==0:
        return(data1[n//2])
    else:
        return((data1[n//2]+data1[n//2-1])/2)
def mode(data1):
    count_dic = {}
    data1.sort()
    prev = data1[0]
    max_count = 0
    count_dic[prev] = 1
    for i in range(1,len(data1)):
        if prev==data1[i]:
            count_dic[data1[i]]+=1
        else:
            count_dic[data1[i]]=1
            if max_count<count_dic[prev]:
                max_count = count_dic[prev]
        prev = data1[i]
    keys = list(count_dic.keys()) 
    mode = []
    for i in keys:
        if count_dic[i]==max_count:
            mode.append(i)
    return mode
```

```python
#loading data
data1 = np.loadtxt("./data1.txt")
data2 = np.loadtxt("./data2.txt")
data3 = np.loadtxt("./data3.txt")
data4 = np.loadtxt("./data4.txt")

data1_median = median(data1)
data1_mode = mode(data1)
data1_mean = mean(data1)
data1_variance = variance(data1)
data1_stddev = stddev(data1)
print("For data1\nmean:{}\nmedian:{}\nvariance:{}\n".format(data1_mean,data1_median,data1_variance))

data2_median = median(data2)
data2_mode = mode(data2)
data2_mean = mean(data2)
data2_variance = variance(data2)
data2_stddev = stddev(data2)
print("For data2\nmean:{}\nmedian:{}\nvariance:{}\n".format(data2_mean,data2_median,data2_variance))


data3_median = median(data3)
data3_mode = mode(data3)
data3_mean = mean(data3)
data3_variance = variance(data3)
data3_stddev = stddev(data3)
print("For data3\nmean:{}\nmedian:{}\nvariance:{}\n".format(data3_mean,data3_median,data3_variance))


data4_median = median(data4)
data4_mode = mode(data4)
data4_mean = mean(data4)
data4_variance = variance(data4)
data4_stddev = stddev(data4)
print("For data4\nmean:{}\nmedian:{}\nvariance:{}\n".format(data4_mean,data4_median,data4_variance))

```
